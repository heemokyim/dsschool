# Otto Group Product Classification Challenge

```python
import pandas as pd
```

## Load Dataset

```python
train = pd.read_csv("data/train.csv", index_col="id")

print(train.shape)
train.head()
```

```python
test = pd.read_csv("data/test.csv", index_col="id")

print(test.shape)
test.head()
```

## Train

```python
label_name = "target"
label_name
```

```python
feature_names = train.columns.difference([label_name])
print(len(feature_names))
feature_names
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
y_train.value_counts()
```

## Use LightGBM

```python
from lightgbm import LGBMClassifier

random_state = 42

model = LGBMClassifier(boosting_type='gbdt',
                       n_estimators=1000,
                       learning_rate=0.017,
                       num_leaves=130,
                       max_bin=100,
                       min_child_samples=60,
                       min_split_gain=0.0,
                       subsample=0.8,
                       subsample_freq=1,
                       colsample_bytree=0.7,
                       n_jobs=-1,
                       random_state=random_state)
model
```

## Evaluate

```python
from sklearn.model_selection import train_test_split

X_train_hold, X_test_hold, y_train_hold, y_test_hold = train_test_split(X_train, y_train, test_size=0.25, random_state=42)

print(X_train_hold.shape, y_train_hold.shape)
print(X_test_hold.shape, y_test_hold.shape)
```

```python
%time model.fit(X_train_hold, y_train_hold)
```

```python
y_test_predict = model.predict_proba(X_test_hold)

print(y_test_predict.shape)
y_test_predict
```

#### understanding log loss
https://datawookie.netlify.com/blog/2015/12/making-sense-of-logarithmic-loss/

```python
from sklearn.metrics import log_loss

score = log_loss(y_test_hold, y_test_predict)
f"Score = {score:.5f}"
```

## Predict

```python
%time model.fit(X_train, y_train)
```

```python
predictions = model.predict_proba(X_test)

print(predictions.shape)
predictions
```

## Submit

```python
sample_submission = pd.read_csv("data/sampleSubmission.csv", index_col="id")

print(sample_submission.shape)
sample_submission.head()
```

```python
submission = pd.DataFrame(predictions,
                         index=X_test.index,
                         columns=model.classes_)

print(submission.shape)
submission.head()
```

```python
submission.to_csv("data/baseline-script-0.42780.csv")
```
